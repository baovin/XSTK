```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # Import a continuous distribution

# Define the continuous random variable
X = norm(loc=0, scale=1)  # Example: standard normal distribution (mean=0, std=1)

# Define the range of values
chuan = np.arange(-4, 4, 0.01)

# Calculate the PDF
pdf = X.pdf(chuan)

# Plot the PDF
plt.plot(chuan, pdf)
plt.xlabel("x")
plt.ylabel("P_X(x)")
plt.title("Probability Density Function")
plt.show()
```

```python

# Calculate the PDF
cdf = X.cdf(chuan)

# Plot the PDF
plt.plot(chuan, cdf)
plt.xlabel("x")
plt.ylabel("P_X(x)")
plt.title("Probability Density Function")
plt.show()
```
