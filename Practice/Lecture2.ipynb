```python
import numpy as np 
```

```python
diem = np.random.randint(0,11,100)
# init 100 random value in range of [0,10]
random_val, frequen = np.unique(diem, return_counts=True)

pmf = frequen / len(diem)

#Table of distribution 

np.column_stack((random_val, pmf))
```

# Chart of weight function


```python
import seaborn as sns
import matplotlib.pyplot as plt

PMF = sns.barplot(pmf)
PMF.set(xlabel='Grade', ylabel='P_X(x)')
plt.show()
```

```python
cdf = np.cumsum(pmf)
CDF = sns.barplot(cdf, label= random_val)
CDF.set(xlabel='Grade', ylabel='F_X(x)')
plt.grid('--', axis='y')
plt.show()
```

```python
# probability >= 4 
1-cdf[4]
```

```python
cdf[7]-cdf[5]
```

# Density function


```python
from scipy.stats import norm 
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4,4,0.001)
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
x = np.arange(-4,4,0.001)
plt.plot(x, norm.cdf(x))
plt.show()
```

```python
cdf = np.cumsum(norm.pdf(x))

plt.plot(x, cdf)
```

```python
# xac suat tai lan can diem 1.5
norm.pdf(1.5)
```

```python
# xac suat tu 1 den 1.5

norm.cdf(1.5) - norm.cdf(1)
```

 ### Biến ngẫu nhiên X có phân bố mũ với hàm mật độ xác suất cho bởi f(x,0)= $\theta e^{-\theta x}, x>0$ trong đó $\theta$ là một tham số dương
1. Tính xác suất để biến ngẫu nhiên nhận giá trị nhỏ hơn 1
2. Tìm và vẽ hàm phân phối xác suất CDF của X
3. Tìm giá trị trung bình, phương sai của X

```python
def pdf(x,theta):
    return np.exp(-theta * x)

x = np.linspace(0, 30, 100)
theta = 1

plt.plot(x, pdf(x, theta))
plt.xlabel('Giá trị của X')
plt.ylabel('PDF')
plt.title('Hàm phân phối xác suất PDF của biến ngẫu nhiên X')
plt.grid(True)
plt.show()





```

```python
import numpy as np
import matplotlib.pyplot as plt



def cdf(x, theta):
    return 1 - np.exp(-theta * x)

x = np.linspace(0, 30, 100)
theta = 1

plt.plot(x, cdf(x, theta))
plt.xlabel('Giá trị của X')
plt.ylabel('CDF')
plt.title('Hàm phân phối xác suất CDF của biến ngẫu nhiên X')
plt.grid(True)
plt.show()
```
