Bài tập thực hành 1
1. Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n = 50, p = 0.7
2. Tính giá trị thống kê min, max, Q1, Q2, Q3
3. Vẽ đồ thị boxplot
4. Vẽ đồ thị tần số, tần suất, hàm mật độ có histogram, chỉ có hàm mật độ

```python
import numpy as np

# Tham số cho phân phối nhị thức
n = 50  # số lần thử
p = 0.7  # xác suất thành công trong mỗi lần thử

# Tạo 5000 số ngẫu nhiên
random_values = np.random.binomial(n, p, 5000)

print(random_values)
```

```python
import numpy as np

# Tính các giá trị thống kê
min_value = np.min(random_values)
max_value = np.max(random_values)
q1 = np.percentile(random_values, 25)
q2 = np.median(random_values)
q3 = np.percentile(random_values, 75)

print(f"Min: {min_value}, Max: {max_value}, Q1: {q1}, Q2: {q2}, Q3: {q3}")
```

```python
import matplotlib.pyplot as plt

# Vẽ đồ thị boxplot
plt.boxplot(random_values)
plt.title('Boxplot của các giá trị ngẫu nhiên phân phối nhị thức')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.show()
```

```python
import seaborn as sns

# Vẽ đồ thị histogram và hàm mật độ
sns.histplot(random_values, kde=True)
plt.title('Histogram và Hàm mật độ của các giá trị ngẫu nhiên phân phối nhị thức')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.show()

# Vẽ chỉ hàm mật độ
sns.kdeplot(random_values, shade=True)
plt.title('Hàm mật độ của các giá trị ngẫu nhiên phân phối nhị thức')
plt.xlabel('Giá trị')
plt.ylabel('Mật độ')
plt.show()
```
